```python
df1 #just printing out df1 to see how it will look like for reference of order
```

```python
dfshuffle = df1.sample(frac=1).reset_index(drop=True)  
# The frac keyword argument specifies the fraction of rows to return in the random sample, 
# so frac=1 means return all rows (in random order).
# ALSO: drop=True prevents .reset_index from creating a column containing the old index entries.

# print shuffle <- the line of code I sometimes used to check and make sure it was shuffling the Students

A = 5 # A is the variable that you can change to any number of groups that you desire to have 
np.array_split(dfshuffle, A) #splits randyrows into 'A' amount of groups and displays them (not pretty)
```
